## Variant 6
$$
\begin{cases}
    -10.3x_1 + 10.2x_2 = 70 \\
    4.7x_1 + 12.3x_2 = 173 \\
    13.2x_1 + 8.8x_2 = 282 \\
\end{cases}
$$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
eq1_coeffs = [-10.3, 10.2, 70]
eq2_coeffs = [4.7, 12.3, 173]
eq3_coeffs = [13.2, 8.8, 282]

def line(eq_coeffs, x):
    return (-eq_coeffs[0] * x + eq_coeffs[2]) / eq_coeffs[1]

x = np.linspace(0, 21, 20)

fig, axs = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

axs[0].plot(x, line(eq1_coeffs, x), color='purple', linestyle='dashed')
axs[1].plot(x, line(eq2_coeffs, x), color='gold', linestyle='dotted')
axs[2].plot(x, line(eq3_coeffs, x), color='red', linestyle='dashdot')

axs[0].set_title(r'â€“10.3$x_1$ + 10.2$x_2$ = 70')
axs[1].set_title(r'4.7$x_1$ + 12.3$x_2$ = 173')
axs[2].set_title(r'13.2$x_1$ + 8.8$x_2$ = 282')

for ax in axs:
    ax.grid(True, color='green')

plt.tight_layout()

plt.savefig('plot1.png', format='png')
plt.savefig('plot1.jpg', format='jpg')
plt.savefig('plot1.svg', format='svg')

plt.show()
```

```python
x = np.linspace(0, 20, 100)

fig, ax = plt.subplots(figsize=(8, 16), dpi=100)

ax.plot(x, line(eq1_coeffs, x), color='purple', linestyle='dashed', label='f(0)')
ax.plot(x, line(eq2_coeffs, x), color='gold', linestyle='dotted', label='f(1)')
ax.plot(x, line(eq3_coeffs, x), color='red', linestyle='dashdot', label='f(2)')

intersection_point1 = np.linalg.solve([eq1_coeffs[:-1], eq2_coeffs[:-1]], [eq1_coeffs[-1], eq2_coeffs[-1]])
intersection_point2 = np.linalg.solve([eq2_coeffs[:-1], eq3_coeffs[:-1]], [eq2_coeffs[-1], eq3_coeffs[-1]])
intersection_point3 = np.linalg.solve([eq1_coeffs[:-1], eq3_coeffs[:-1]], [eq1_coeffs[-1], eq3_coeffs[-1]])

fill_vertices = np.array([intersection_point1, intersection_point2, intersection_point3, intersection_point1])
ax.fill(fill_vertices[:, 0], fill_vertices[:, 1], color='lightgray', alpha=0.5)

ax.scatter(*intersection_point1, color='red', marker='o')
ax.scatter(*intersection_point2, color='green', marker='o')
ax.scatter(*intersection_point3, color='blue', marker='o')

ax.annotate(f'Point of intersection of f(0) and f(1)', intersection_point1, textcoords="offset points", xytext=(40,-140), ha='center', arrowprops=dict(facecolor='black', arrowstyle='->'))
ax.annotate(f'Point of intersection of f(1) and f(2)', intersection_point2, textcoords="offset points", xytext=(10,180), ha='center', arrowprops=dict(facecolor='black', arrowstyle='->'))
ax.annotate(f'Point of intersection of f(0) and f(2)', intersection_point3, textcoords="offset points", xytext=(20,150), ha='center', arrowprops=dict(facecolor='black', arrowstyle='->'))

ax.text(5, line(eq1_coeffs, 6.5), r'â€“10.3$x_1$ + 10.2$x_2$ = 70', color='purple', fontsize=8, rotation=np.degrees(np.arctan(10.3/10.2)))
ax.text(7, line(eq2_coeffs, 9), r'4.7$x_1$ + 12.3$x_2$ = 173', color='gold', fontsize=8, rotation=np.degrees(np.arctan(-4.7/12.3)))
ax.text(13, line(eq3_coeffs, 14), r'13.2$x_1$ + 8.8$x_2$ = 282', color='red', fontsize=8, rotation=np.degrees(np.arctan(-13.2/8.8)))

ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_title('All lines in one figure')

ax.grid(True, color='green', linestyle='dotted')    

ax.legend()

plt.savefig('plot2.png', format='png')
plt.savefig('plot2.jpg', format='jpg')
plt.savefig('plot2.svg', format='svg')

plt.show()
```
